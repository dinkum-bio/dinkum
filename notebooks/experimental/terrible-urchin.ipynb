```python
class SeaUrchin_Blastula:
    macromere1 = [(120, 250),
                  (220, 250),
                  (220, 350),
                  (120, 350) ]
    macromere2 = [(320, 270),
                  (420, 270),
                  (420, 370),
                  (320, 370) ]

    mesomere1 = [(140, 100),
                  (220, 100),
                  (220, 180),
                  (140, 180) ]
    mesomere2 = [(300, 100),
                  (380, 100),
                  (380, 180),
                  (300, 180) ]

    micromere1 = [(195, 410),
                  (210, 410),
                  (210, 425),
                  (195, 425) ]
    micromere2 = [(250, 415),
                  (265, 415),
                  (265, 430),
                  (250, 430) ]    

    def __init__(self):
        self.load()
        self.image = self.orig_image.copy()
        self.imdraw = ImageDraw.Draw(self.image)
    
    def load(self):
        orig_im = Image.open('../urchin-embryo.png')
        #print(orig_im.format, orig_im.size, orig_im.mode)

        orig_im = orig_im.resize((512, 512))
        #print(orig_im.format, orig_im.size, orig_im.mode)
        self.orig_image = orig_im
        
    def fill(self, cell_type, color):
        assert cell_type in ('macromere1', 'micromere1', 'mesomere1',
                             'macromere2', 'micromere2', 'mesomere2')
        
        polygon_coords = getattr(self, cell_type)
        self.imdraw.polygon(polygon_coords, fill=color)
        
    
```

```python
blastula = SeaUrchin_Blastula()

blastula.fill('macromere1', "red")
blastula.fill('macromere2', 'red')
blastula.fill('mesomere1', 'cyan')
blastula.fill('mesomere2', 'cyan')

blastula.fill('micromere1', 'green')
blastula.fill('micromere2', 'orange')

blastula.image
```
